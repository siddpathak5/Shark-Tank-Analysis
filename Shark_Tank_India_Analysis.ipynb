# **2. Data Cleaning:**

```python
#viewing all the columns.
Shark_Tank.columns
```

```python
#Identifying the columns with missing values.
missing_values = Shark_Tank.isna().sum()

#finding the total percentage of missing values.
missing_percentage = (missing_values/len(Shark_Tank))*100
missing_data = pd.DataFrame({'Missing Values':missing_values,'Missing Percentage':missing_percentage}).loc[missing_values>0]
missing_data
```

```python
#Removing unwanted column from the dataset.
Shark_Tank.drop(columns= ['Past valuation ', 'Unnamed: 23', 'Unnamed: 24'], inplace = True)
```

```python
# Dropping rows with missing values in 'Revenue YoY', 'Ask_value', and 'Ask_% of Equity' columns.
Shark_Tank.dropna(subset=['Revenue YoY', 'Ask_value', 'Ask_% of Equity'], inplace=True)

# Impute missing values in 'Valuation after Deal' with the median of the column
Shark_Tank['Valuation after Deal'].fillna(Shark_Tank['Valuation after Deal'].median(), inplace=True)
```

```python
#Rechecking for null values.
Shark_Tank.isna().sum()
```

# **3. Data Preprocessing:**

```python
#Identifying and removing outliers from the numerical columns.

# List of numerical columns to check for outliers.
numeric_columns = ['No. of Pitchers', 'Mean Founder Age', 'Revenue YoY', 'Ask_value', 'Deal_value', 'Valuation after Deal']

# Iterate over each column and remove rows with outliers
for column in numeric_columns:
    # Calculate the first quartile (Q1) and third quartile (Q3)
    Q1 = Shark_Tank[column].quantile(0.25)
    Q3 = Shark_Tank[column].quantile(0.75)

    # Calculate the Interquartile Range (IQR)
    IQR = Q3 - Q1

    # Define lower and upper bounds for detecting outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Filter out rows that are within the bounds
    Shark_Tank = Shark_Tank[(Shark_Tank[column] >= lower_bound) & (Shark_Tank[column] <= upper_bound)]

# Plot boxplots for each column to visually inspect after outlier removal
Shark_Tank[numeric_columns].boxplot(figsize=(10, 6))
plt.show()
```

# **4. Exploratory Data Analysis (EDA) and Visualization:**

# **Objective of the Analysis**

The objective of this analysis is to explore various factors influencing startup funding outcomes, such as sector preferences, founder demographics, equity and valuation dynamics, and investor behavior. By investigating these aspects, we aim to understand which characteristics of startups and their founders attract investment, how different sectors fare in securing deals, the impact of financial requests and offered equity on deal outcomes, and the strategic preferences of individual investors. This analysis provides insights into the dynamics of startup funding, helping entrepreneurs and investors make informed decisions.


```python
#Q1. Which Sectors Attracted the Most Deals?

# Filter the DataFrame to include only rows where a deal was made
successful_deals = Shark_Tank[Shark_Tank['Deal'] == 'Y']

# Group by 'Sector' and count the number of deals in each sector
deals_per_sector = successful_deals['Sector'].value_counts().reset_index()

# Rename columns for better understanding
deals_per_sector.columns = ['Sector', 'Number of Deals']

# Sort the sectors by the number of deals in descending order
deals_per_sector.sort_values(by='Number of Deals', ascending=False, inplace=True)

# Plotting the results using Matplotlib
plt.figure(figsize=(10, 6))
plt.bar(deals_per_sector['Sector'], deals_per_sector['Number of Deals'], color='skyblue')

# Adding title and labels
plt.title('Number of Deals by Sector')
plt.xlabel('Sector')
plt.ylabel('Number of Deals')

# Adding data labels on top of each bar
for index, value in enumerate(deals_per_sector['Number of Deals']):
    plt.text(index, value + 0.05, str(value), ha='center')

# Rotating x-axis labels for better readability
plt.xticks(rotation=90)

# Display the plot
plt.show()
```

```python
#Q2. Show comparison of ask value vs deal value?
# Filter the DataFrame to include only rows with non-zero Ask_value and Deal_value
df_filtered = Shark_Tank[(Shark_Tank['Ask_value'] > 0) & (Shark_Tank['Deal_value'] > 0)]

# Plotting the bar plot
plt.figure(figsize=(10, 6))
bar_width = 0.35
index = range(len(df_filtered))

# Creating bars for Ask_value and Deal_value
plt.bar(index, df_filtered['Ask_value'], width=bar_width, label='Ask Value', color='blue', alpha=0.6)
plt.bar([i + bar_width for i in index], df_filtered['Deal_value'], width=bar_width, label='Deal Value', color='green', alpha=0.6)

# Adding labels and title
plt.xlabel('Start-Up Name')
plt.ylabel('Value (in lakhs)')
plt.title('Comparison of Ask Value and Deal Value for Start-Ups with Non-Zero Values')
plt.xticks([i + bar_width/2 for i in index], df_filtered['Start-Up Name'], rotation=90)
plt.legend()

# Display the plot
plt.tight_layout()
plt.show()
```

```python
#Q3. How does the percentage of equity asked by the founders (Ask_% of Equity) impact the likelihood of getting a deal?
# Calculate deal probability by equity percentage
deal_probability = Shark_Tank.groupby('Ask_% of Equity')['Deal'].apply(lambda x: (x == 'Y').mean()).reset_index()

# Rename columns for clarity
deal_probability.columns = ['Ask_% of Equity', 'Probability of Deal']

# Plotting the relationship between 'Ask_% of Equity' and the 'Probability of Deal'
plt.figure(figsize=(10, 6))
plt.plot(deal_probability['Ask_% of Equity'], deal_probability['Probability of Deal'], marker='o', linestyle='-', color='blue')

# Adding title and labels
plt.title('Impact of Equity Percentage on Deals')
plt.xlabel('Ask % of Equity')
plt.ylabel('Probability of Deal')

# Add point labels
for i, row in deal_probability.iterrows():
    plt.text(row['Ask_% of Equity'], row['Probability of Deal'], f'{row["Probability of Deal"]:.2f}',
             ha='center', va='bottom', fontsize=9, color='black')

# Display the plot
plt.grid()
plt.show()
```

```python
#Q4. What is the distribution of founder ages among startups that received deals?
# Group by 'Mean Founder Age' and count the number of successful pitches
distribution = Shark_Tank.groupby('Mean Founder Age')['Deal'].apply(lambda x: (x == 'Y').sum()).reset_index()

# Rename columns
distribution.columns = ['Mean Founder Age', 'Successful Pitches']
plt.figure(figsize=(10, 6))
plt.plot(distribution['Mean Founder Age'], distribution['Successful Pitches'], marker='o', linestyle='-', color='blue')

# Adding title and labels
plt.title('Distribution of Founder Ages in Successful Pitches')
plt.xlabel('Founder Ages')
plt.ylabel('Successful Pitches')

# Add point labels
for i, row in distribution.iterrows():
    plt.text(row['Mean Founder Age'], row['Successful Pitches'], f'{row["Successful Pitches"]:.2f}',
             ha='center', va='bottom', fontsize=9, color='black')

# Display the plot
plt.grid()
plt.show()
```

```python
#Q5. Does the gender of the founders (as indicated by Sex) impact the likelihood of securing a deal?
# Group by 'Sex' and count the number of successful pitches
Gender_dynamics = Shark_Tank.groupby('Sex')['Deal'].apply(lambda x: (x == 'Y').sum()).reset_index()

# Rename columns
Gender_dynamics.columns = ['Gender', 'Deal Count']

# Plotting
plt.figure(figsize=(8, 5))
bars = plt.bar(Gender_dynamics['Gender'], Gender_dynamics['Deal Count'], color='Red')

# Add labels and title
plt.xlabel('Gender')
plt.ylabel('Successful Deals Count')
plt.title('Number of Successful Deals by Gender')

# Adding labels on bars
for bar in bars:
    yval = bar.get_height()  # Get the height of the bar
    plt.text(bar.get_x() + bar.get_width()/2, yval, int(yval), ha='center', va='bottom', color='black', fontsize=10)

# Show the plot
plt.tight_layout()  # Adjust layout to fit labels
plt.show()
```

```python
#Q6. Which investors prefer to invest in specific sectors or types of ideas?
investor_sector_counts = Shark_Tank.melt(id_vars=['Sector'], value_vars=['Ashneer Grover', 'Vineeta Singh', 'Peyush Bansal', 'Namita Thapar', 'Anupam Mittal', 'Ghazal Alagh'],
                                         var_name='Investor', value_name='Investment').query('Investment > 0')

# Pivot the data to create a matrix of investors and sectors
investor_sector_matrix = investor_sector_counts.pivot_table(index='Sector', columns='Investor', values='Investment', aggfunc='count', fill_value=0)

# Plotting the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(investor_sector_matrix, annot=True, cmap='Blues')
plt.title('Investor Preferences by Sector')
plt.show()
```

```python
#Q7. Is there a relationship between Revenue YoY and the Valuation after Deal?
# Calculate correlation
revenue_valuation_corr = Shark_Tank['Revenue YoY'].corr(Shark_Tank['Valuation after Deal'])
print(f'Correlation between Revenue YoY and Valuation after Deal: {revenue_valuation_corr:.2f}')

# Prepare data for heatmap
corr_matrix = Shark_Tank[['Revenue YoY', 'Valuation after Deal']].corr()

# Create subplots
fig, axs = plt.subplots(1, 2, figsize=(16, 6))

# Scatter plot
axs[0].scatter(Shark_Tank['Revenue YoY'], Shark_Tank['Valuation after Deal'], color='orange', alpha=0.6)
axs[0].set_title('Scatter Plot of Revenue YoY vs. Valuation After Deal')
axs[0].set_xlabel('Revenue YoY')
axs[0].set_ylabel('Valuation After Deal')

# Heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, center=0, fmt='.2f', linewidths=0.5, ax=axs[1])
axs[1].set_title('Correlation Heatmap of Revenue YoY vs. Valuation After Deal')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
#Q8. How common are loan deals, and which sectors or startups are more likely to involve loans?
# Filter startups with loan deals
loan_deals = Shark_Tank[Shark_Tank['Deal Loan'] > 0]

# Count loan deals by sector
loan_deals_per_sector = loan_deals['Sector'].value_counts()

# Plotting the results
plt.figure(figsize=(10, 6))
bars = plt.bar(loan_deals_per_sector.index, loan_deals_per_sector.values, color='purple')
plt.title('Distribution of Loan Deals by Sector')
plt.xlabel('Sector')
plt.ylabel('Number of Loan Deals')
plt.xticks(rotation=45)

# Adding labels on bars
for bar in bars:
    yval = bar.get_height()  # Get the height of the bar
    plt.text(bar.get_x() + bar.get_width()/2, yval, int(yval), ha='center', va='bottom', color='black', fontsize=10)

plt.show()
```

```python
# Removing any leading or trailing whitespace from all column names.
Shark_Tank.columns = Shark_Tank.columns.str.strip()

# Converting 'Deal_% of Equity' & 'Ask_% of Equity' from a string with a '%' sign to a float.
# 1. Removing the '%' sign from each value using .str.rstrip('%')
# 2. Converting the resulting string to a float data type with .astype('float')
Shark_Tank['Deal_% of Equity'] = Shark_Tank['Deal_% of Equity'].str.rstrip('%').astype('float')
Shark_Tank['Ask_% of Equity'] = Shark_Tank['Ask_% of Equity'].str.rstrip('%').astype('float')
```

```python
#Q9. How does the equity percentage offered (Deal_% of Equity) vary across different sectors?
# Calculate the average Deal_% of Equity for each sector
equity_by_sector = Shark_Tank.groupby('Sector')['Deal_% of Equity'].mean().reset_index()

# Sort by equity percentage
equity_by_sector.sort_values(by='Deal_% of Equity', ascending=False, inplace=True)

# Plotting the results
plt.figure(figsize=(12, 6))
bars = plt.bar(equity_by_sector['Sector'], equity_by_sector['Deal_% of Equity'], color='lightgreen')
plt.xticks(rotation=45, ha='right')
plt.title('Average Equity Percentage Offered by Sector')
plt.xlabel('Sector')
plt.ylabel('Average Deal % of Equity')

# Adding labels on bars
for bar in bars:
    yval = bar.get_height()  # Get the height of the bar
    plt.text(bar.get_x() + bar.get_width()/2, yval, f'{yval:.2f}', ha='center', va='bottom', color='black', fontsize=10)

plt.show()
```

```python
#Q10. Which startups received funding from multiple investors, and is there a pattern in these syndicates?
# Count the number of non-zero investor columns for each startup
investor_columns = ['Aman Gupta', 'Ashneer Grover', 'Vineeta Singh', 'Peyush Bansal', 'Namita Thapar', 'Anupam Mittal', 'Ghazal Alagh']
Shark_Tank['Number of Investors'] = Shark_Tank[investor_columns].astype(bool).sum(axis=1)

# Filter startups with more than one investor
startups_with_multiple_investors = Shark_Tank[Shark_Tank['Number of Investors'] > 1]

# Plotting the results
plt.figure(figsize=(12, 6))
bars = plt.bar(startups_with_multiple_investors['Start-Up Name'], startups_with_multiple_investors['Number of Investors'], color='lightblue')
plt.xticks(rotation=45, ha='right')
plt.title('Number of Investors for Each Startup')
plt.xlabel('Start-Up Name')
plt.ylabel('Number of Investors')

# Adding labels on bars
for bar in bars:
    yval = bar.get_height()  # Get the height of the bar
    plt.text(bar.get_x() + bar.get_width()/2, yval, int(yval), ha='center', va='bottom', color='black', fontsize=10)

plt.show()
```

```python
#Q11. How does the Valuation after Deal compare to the initial Ask_value?
# Calculate valuation multiple
Shark_Tank['Valuation Multiple'] = Shark_Tank['Valuation after Deal'] / Shark_Tank['Ask_value']

# Create a figure with subplots
fig, ax = plt.subplots(1, 2, figsize=(18, 6))

# Subplot 1: Distribution of valuation multiples
sns.histplot(Shark_Tank['Valuation Multiple'], bins=20, kde=True, color='purple', ax=ax[0])
ax[0].set_title('Distribution of Valuation Multiples')
ax[0].set_xlabel('Valuation Multiple')
ax[0].set_ylabel('Frequency')

# Analyze valuation multiples by sector
valuation_by_sector = Shark_Tank.groupby('Sector')['Valuation Multiple'].mean().reset_index()
valuation_by_sector.sort_values(by='Valuation Multiple', ascending=False, inplace=True)

# Subplot 2: Average valuation multiples by sector
bars = ax[1].bar(valuation_by_sector['Sector'], valuation_by_sector['Valuation Multiple'], color='dodgerblue')
ax[1].set_title('Average Valuation Multiple by Sector')
ax[1].set_xlabel('Sector')
ax[1].set_ylabel('Average Valuation Multiple')
ax[1].tick_params(axis='x', rotation=90)

# Add bar labels
for bar in bars:
    yval = bar.get_height()
    ax[1].text(bar.get_x() + bar.get_width() / 2, yval, round(yval, 2), ha='center', va='bottom')

# Adjust layout and show plot
plt.tight_layout()
plt.show()
```

```python
#Q12. What is the distribution of deal sizes (Deal_value) across different sectors?
# Calculate the average deal value for each sector
avg_deal_value_per_sector = Shark_Tank.groupby('Sector')['Deal_value'].agg(['mean']).reset_index()
avg_deal_value_per_sector.columns = ['Sector', 'Average Deal Value']

plt.figure(figsize=(12, 8))

# Create the bar plot
plt.bar(avg_deal_value_per_sector['Sector'], avg_deal_value_per_sector['Average Deal Value'], color='yellow')

# Add labels and formatting
plt.xticks(rotation=45, ha='right')
plt.title('Distribution of Deal Sizes Across Sectors')
plt.xlabel('Sector')
plt.ylabel('Average Deal Value (in Lakhs)')

# Adding labels on bars
for index, row in avg_deal_value_per_sector.iterrows():
    plt.text(index, row['Average Deal Value'] + 1, f'{row["Average Deal Value"]:.2f}', ha='center', va='bottom', fontsize=10)

plt.show()
```

```python
#Q13. What is the total number of accepted and rejected deals?
rejected_deals = Shark_Tank['Deal'].value_counts().reset_index()
rejected_deals.columns = ['Deal Status', 'Count']

# Plotting the bar plot
plt.figure(figsize=(6, 4))
bars = plt.bar(x=rejected_deals['Deal Status'], height=rejected_deals['Count'], color='skyblue')

# Add labels and title
plt.title('Accepted vs. Rejected Deals')
plt.xlabel('Deal Status (Y = Accepted, N = Rejected)')
plt.ylabel('Number of Deals')
for bar in bars:
  yval = bar.get_height()
  plt.text(bar.get_x() + bar.get_width()/2, yval, int(yval), ha='center', va='bottom', color='black', fontsize=10)

# Display the plot
plt.show()
```

```python
#Q14. What are the sectors that have maximum rejection rate of deals?
# Filter startups with no deals
rejected_startups = Shark_Tank[Shark_Tank['Deal'] == 'N']

# Analyze sectors of rejected startups
rejection_by_sector = rejected_startups['Sector'].value_counts()

# Plotting the results
plt.figure(figsize=(10, 6))
bars = plt.bar(rejection_by_sector.index, rejection_by_sector.values, color='red')
plt.title('Sectors with Highest Rejection Rates')
plt.xlabel('Sector')
plt.ylabel('Number of Rejections')
plt.xticks(rotation=90)
for bar in bars:
    yval = bar.get_height()  # Get the height of the bar
    plt.text(bar.get_x() + bar.get_width()/2, yval, int(yval), ha='center', va='bottom', color='black', fontsize=10)
plt.show()
```

```python
#Q15. What are the sectors that have maximum acceptance rate of deals?
# Filter startups with no deals
accepted_startups = Shark_Tank[Shark_Tank['Deal'] == 'Y']

# Analyze sectors of rejected startups
acceptance_by_sector = accepted_startups['Sector'].value_counts()

# Plotting the results
plt.figure(figsize=(10, 6))
bars = plt.bar(rejection_by_sector.index, rejection_by_sector.values, color='green')
plt.title('Sectors with Highest Acceptance Rates')
plt.xlabel('Sector')
plt.ylabel('Number of Acceptance')
plt.xticks(rotation=90)
for bar in bars:
    yval = bar.get_height()  # Get the height of the bar
    plt.text(bar.get_x() + bar.get_width()/2, yval, int(yval), ha='center', va='bottom', color='black', fontsize=10)
plt.show()
```

```python
#Q16. How diverse are the investors in terms of the sectors they invest in?
# Set up the subplot grid
fig, axes = plt.subplots(3, 3, figsize=(18, 15))  # 3x3 grid
axes = axes.flatten()  # Flatten the 2D array of axes for easy iteration

# Iterate through each investor and plot
for i, investor in enumerate(investor_columns):
    investor_sector_counts = Shark_Tank[Shark_Tank[investor] > 0]['Sector'].value_counts()

    # Plotting a pie chart for each investor
    axes[i].pie(investor_sector_counts.values, labels=investor_sector_counts.index, autopct='%1.1f%%', startangle=140)
    axes[i].set_title(f'Sector Diversity of Investments by {investor}')
    axes[i].axis('equal')

# Remove the extra subplot (if there are fewer than 9 plots)
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python
#Q17. What is the total deal value and average equity percentage by sector?
sector_analysis = Shark_Tank.groupby('Sector').agg({'Deal_value': 'sum','Deal_% of Equity': 'mean'}).reset_index()

# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 8))

# Plotting total deal value by sector
bars1 = sns.barplot(x='Sector', y='Deal_value', data=sector_analysis, palette='viridis', ax=ax1)
ax1.set_title('Total Deal Value by Sector')
ax1.set_xlabel('Sector')
ax1.set_ylabel('Total Deal Value (in Lakhs)')
ax1.tick_params(axis='x', rotation=45)

# Add bar labels for total deal value
for bar in bars1.patches:
    yval = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2, yval, f'{yval:,.0f}', ha='center', va='bottom', color='black', fontsize=10)

# Plotting average equity percentage by sector
bars2 = sns.barplot(x='Sector', y='Deal_% of Equity', data=sector_analysis, palette='coolwarm', ax=ax2)
ax2.set_title('Average Deal % of Equity by Sector')
ax2.set_xlabel('Sector')
ax2.set_ylabel('Average Deal % of Equity')
ax2.tick_params(axis='x', rotation=45)

# Add bar labels for average equity percentage
for bar in bars2.patches:
    yval = bar.get_height()
    ax2.text(bar.get_x() + bar.get_width()/2, yval, f'{yval:.2f}%', ha='center', va='bottom', color='black', fontsize=10)

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```

```python
#Q18. Find the top 20 deals by 'Valuation after Deal' across IDEA? (#IDEA stands for "Innovative Design, Engineering, and Architecture")
Top_20_Deals = Shark_Tank.nlargest(20, 'Valuation after Deal')

# Plotting top 20 deals with their 'IDEA'
plt.figure(figsize=(14, 6))
bar_plot = sns.barplot(x='IDEA', y='Valuation after Deal', data=Top_20_Deals, palette='Spectral')
plt.title('Top 20 Deals by overall Valuation across IDEA')
plt.xlabel('IDEA')
plt.ylabel('Valuation (in lakhs)')
plt.xticks(rotation=90)

# Add bar labels
for bar in bar_plot.patches:
    yval = bar.get_height()
    bar_plot.text(
        bar.get_x() + bar.get_width() / 2,yval,f'{int(yval):,}',ha='center',va='bottom',fontsize=10,color='black')

plt.show()
```

```python
#Q19. Calculate the average Revenue YoY for each Sector?
avg_revenue_per_sector = Shark_Tank.groupby('Sector')['Revenue YoY'].mean().reset_index()

# Sorting by Revenue YoY for better visualization
avg_revenue_per_sector.sort_values(by='Revenue YoY', ascending=False, inplace=True)

# Plotting
plt.figure(figsize=(12, 6))
bars = plt.bar(avg_revenue_per_sector['Sector'], avg_revenue_per_sector['Revenue YoY'], color='skyblue')
plt.xlabel('Sector')
plt.ylabel('Average Revenue YoY (in Lakhs)')
plt.title('Average Revenue YoY for Sectors with Deals')
plt.xticks(rotation=90)

# Adding labels on top of the bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval, int(yval), ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
#Q20. Which investor contributed the most in terms of deal value, and what percentage of the total deal value does this represent?
#storing all the investors in a vairable
investors = ['Aman Gupta', 'Ashneer Grover', 'Vineeta Singh', 'Peyush Bansal', 'Namita Thapar', 'Anupam Mittal', 'Ghazal Alagh']
investor_contributions = {}

for investor in investors:
    investor_contributions[investor] = Shark_Tank[investor].sum()

# Convert to DataFrame for easier analysis
investor_contributions_df = pd.DataFrame(list(investor_contributions.items()), columns=['Investor', 'Total Deal Value'])

# Identify the investor with the highest deal value
top_investor = investor_contributions_df.loc[investor_contributions_df['Total Deal Value'].idxmax()]

# Calculate the total deal value
total_deal_value = Shark_Tank['Deal_value'].sum()

# Calculate the percentage contribution
top_investor_percentage = (top_investor['Total Deal Value'] / total_deal_value) * 100

# Plotting the contributions of each investor
plt.figure(figsize=(12, 6))
sns.barplot(x='Investor', y='Total Deal Value', data=investor_contributions_df, palette='viridis')
plt.title('Deal Value Contribution by Investor')
plt.xlabel('Investor')
plt.ylabel('Total Deal Value')
plt.xticks(rotation=45, ha='right')

# Adding labels on top of the bars
for index, row in investor_contributions_df.iterrows():
    plt.text(index, row['Total Deal Value'] + 0.1, f'{row["Total Deal Value"]:.2f}',
             ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```

# **CONCLUSION**
This analysis offers key insights into the factors influencing startup funding, such as sector preferences, founder demographics, and investment dynamics. The Packaged Food sector stands out for its high number of deals and investor interest, while Technology and Lifestyle/Wearables also perform well. Startups with moderate funding requests and strategic equity offers have higher chances of securing deals, particularly within specific equity ranges (2%, 4%, 8%). The analysis also suggests potential gender biases in funding outcomes, with male founders or mixed-gender teams more likely to receive investment. Additionally, revenue growth is not a significant predictor of post-deal valuation, indicating that other factors are more influential. Overall, the findings highlight the importance of strategic positioning and investor behavior understanding for successful funding.
